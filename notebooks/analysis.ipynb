```python
import pandas as pd
```

```python
crop = pd.read_csv("../data/crop_data.csv")
weather = pd.read_csv("../data/weather_data.csv")
crop.head()
weather.head()
```

```python
crop.dropna(inplace=True)
weather.dropna(inplace=True)

crop['Year'] = pd.to_datetime(crop['Year'], format='%Y')
weather['Year'] = pd.to_datetime(weather['Year'], format='%Y')
```

```python
data = pd.merge(crop, weather, on="Year")
data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data['Year'], data['Yield'])
plt.title("Yield Over Years")
plt.show()
```

```python
import seaborn as sns
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
plt.scatter(data['Rainfall'], data['Yield'])
plt.xlabel("Rainfall")
plt.ylabel("Yield")
plt.show()
```
